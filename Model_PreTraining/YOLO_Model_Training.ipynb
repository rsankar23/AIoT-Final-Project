```python
from ultralytics import YOLO

# Load an Open Images Dataset V7 pretrained YOLOv8n model
model = YOLO("yolov8n.pt")

# Run prediction
# results = model.predict(source="image.jpg")

# Start training from the pretrained checkpoint
results = model.train(data="HomeObjects-3K.yaml", epochs=100, imgsz=640, device=[-1,-1], cos_lr=True)
```

```python
test_results = model.val(plots = True, project = "val_res", visualize = True, device = "cuda:0")
```
